### Setup

```python
####################################################################################################
### read and process

# keep an example at hand
lines = """0,9 -> 5,9
8,0 -> 0,8
9,4 -> 3,4
2,2 -> 2,1
7,0 -> 7,4
6,4 -> 2,0
0,9 -> 2,9
3,4 -> 1,4
0,0 -> 8,8
5,5 -> 8,2"""

f = open("5.txt", "r")
lines = f.read()
f.close()

lines = lines.split("\n")
for i, l in enumerate(lines):

    # process start and end points
    p1 = l.split(" ")[0].split(",")
    p1 = tuple(map(int, p1))

    p2 = l.split(" ")[2].split(",")
    p2 = tuple(map(int, p2))

    # modify in place
    lines[i] = {"p1" : p1, "p2" : p2}
    # print(p1, "->", p2)

# further steps of processing: determine orientation, then sort
for l in lines:

    # print(l)

    # horizontal lines have same row indices
    if l["p1"][0] == l["p2"][0]:
        l["orientation"] = "horizontal"

        # put lower column index first
        if l["p1"][1] > l["p2"][1]:
            tmp = l["p1"]
            l["p1"] = l["p2"]
            l["p2"] = tmp

        # put direction for consistency
        l["direction"] = (0, 1)

    # vertical lines have same column indices
    elif l["p1"][1] == l["p2"][1]:
        l["orientation"] = "vertical"

        # put lower row index first
        if l["p1"][0] > l["p2"][0]:
            tmp = l["p1"]
            l["p1"] = l["p2"]
            l["p2"] = tmp
        
        # put direction for consistency
        l["direction"] = (1, 0)

    # else case, diagonal line
    else:
        l["orientation"] = "diagonal"

        # put lower row index first
        if l["p1"][0] > l["p2"][0]:
            tmp = l["p1"]
            l["p1"] = l["p2"]
            l["p2"] = tmp

        # determine orientation: (1, 1) or (1, -1) - first index is 1, because of sorting
        dx = l["p2"][1] - l["p1"][1]
        dy = l["p2"][0] - l["p1"][0]
        l["direction"] = ( 1, int(dy / dx) )

    # print(l)

####################################################################################################
### PART 1

# init board, get size and fill with 0s
rows = 0
cols = 0

for l in lines:

    # p2 row indices are always greater because of sorting
    if rows < l["p2"][0] + 1:
        rows = l["p2"][0] + 1

    # get p1, p2 column indices
    if cols < l["p1"][1] + 1:
        cols = l["p1"][1] + 1
    if cols < l["p2"][1] + 1:
        cols = l["p2"][1] + 1

# print(f"Rows: {rows}, Columns: {cols}")
table = [ [ 0 for c in range(cols) ] for r in range(rows) ]

# consider now only horizontal and vertical lines
for l in lines:

    # horizontal lines increase only in colums
    if l["orientation"] == "horizontal":
        r = l["p1"][0]
        c1 = l["p1"][1]
        c2 = l["p2"][1]

        for c in range(c1, c2 + 1):
            table[r][c] += 1
    
    # vertical orientation, increase only in rows
    if l["orientation"] == "vertical":
        r1 = l["p1"][0]
        r2 = l["p2"][0]
        c = l["p1"][1]

        for r in range(r1, r2 + 1):
            table[r][c] += 1

# check
# for row in table: print(row)

# transpose list to reflect AoC example :)
def transpose(t):
    global rows, cols
    return [ [ t[r][c] for r in range(rows) ] for c in range(cols) ]
table = transpose(table)

# check
# for row in table: print(row)

# answer
c = 0
for row in table:
    c += sum([col > 1 for col in row])
print(f"Answer: {c}")

####################################################################################################
### PART 2

# init board again, no need to recalculate
table = [ [ 0 for c in range(cols) ] for r in range(rows) ]

# now consider diagonals as well
for l in lines:

    # print(l)

    # horizontal lines increase only in colums
    if l["orientation"] == "horizontal":
        r = l["p1"][0]
        c1 = l["p1"][1]
        c2 = l["p2"][1]

        for c in range(c1, c2 + 1):
            table[r][c] += 1
    
    # vertical orientation, increase only in rows
    elif l["orientation"] == "vertical":
        r1 = l["p1"][0]
        r2 = l["p2"][0]
        c = l["p1"][1]

        for r in range(r1, r2 + 1):
            table[r][c] += 1

    # handle diagonals with their saved orientation
    else:
        r1 = l["p1"][0]
        r2 = l["p2"][0]
        c1 = l["p1"][1]
        c2 = l["p2"][1]
        
        # rows are always incremented by 1
        # columns are incremented by inferred direction, thus change the limit as well accordingly
        while r1 != r2 + 1 and c1 != c2 + l["direction"][1]:
            table[r1][c1] +=1
            r1 += 1
            c1 += l["direction"][1]

table = transpose(table)
# for row in table: print(row)

# answer
c = 0
for row in table:
    c += sum([col > 1 for col in row])
print(f"Answer: {c}")
```
