```python
####################################################################################################
### read & process

lines = """11111
19991
19191
19991
11111"""

lines = """5483143223
2745854711
5264556173
6141336146
6357385478
4167524645
2176841721
6882881134
4846848554
5283751526"""

f = open("11.txt")
lines = f.read()
f.close()

lines = lines.split("\n")
lines = [ list(map(int, list(l))) for l in lines ]
lines

####################################################################################################
### PART 1

current = [ l.copy() for l in lines ]

# for c in current: print(c)
# print()

R = len(current)
C = len(current[0])

ans = 0

all_flashed = False

# task 1: execute 100 steps and sum up the total flashes
# for iter in range(100):

iter = 0
# task 2: find that moment where all uoctopses flash simultaneously
while not all_flashed:

    # increment each octopus energy by 1
    for i in range(R):
        for j in range(C):
            current[i][j] += 1
    
    # collect those octopuses which are about to flash
    flashing = [ (i, j) for i in range(R) for j in range(C) if current[i][j] > 9 ]
    # print(flashing)

    flashed = []

    # while is there a octopus which can flash
    while len(flashing) != 0:

        (i, j) = flashing[0]
        # print( "(i, j):", (i, j) )

        # calculate neighbours
        u = i - 1 if i - 1 >= 0 else 0
        d = i + 1 if i + 1 <= R - 1 else R - 1
        l = j - 1 if j - 1 >= 0 else 0
        r = j + 1 if j + 1 <= C - 1 else C - 1
        # print("Box coords:", "u:", u, "d:", d, "l:", l, "r:", r)

        neighbours = [ (u, l), (u, j), (u, r),
                       (i, l),         (i, r),
                       (d, l), (d, j), (d, r) ]

        # remove center point, and remove duplicates
        neighbours = [ n for n in neighbours if n != (i, j) ]
        neighbours = list(set(neighbours))
        
        # print("Neighbours:", neighbours)

        # center octopus flashes, thus increments the energy level of other octopuses
        for n in neighbours:
            (ni, nj) = n
            # print( "(ni, nj):", (ni, nj) )

            # do not increase energy of neighbour, if it is about to flash or already flashed
            if ((ni, nj) not in flashing) and ((ni, nj) not in flashed):
            # if (ni, nj) not in flashed:
                current[ni][nj] += 1

                if current[ni][nj] > 9:
                    flashing.append( (ni, nj) )

        # drop energy of center octopus to 0, move it to flashed
        current[i][j] = 0
        flashing.remove( (i, j) )
        flashed.append( (i, j) )
        ans += 1

    # print(iter + 1)
    # for c in current: print(c)
    # print()

    # task 2: iterate until all octopuses flash at the same time
    all_flashed = (sorted(flashed) == [ (i, j) for i in range(R) for j in range(C) ])

    if all_flashed:
        # print("All flashed during iteration", iter + 1)
        break

    if iter == 1000:
        break
    iter += 1

print("Answer:", ans)

####################################################################################################
### PART 2

ans = iter + 1
print("Answer:", ans)
```
