```python
####################################################################################################
### read & process

lines = """start-A
start-b
A-c
A-b
b-d
A-end
b-end"""

lines = """dc-end
HN-start
start-kj
dc-start
dc-HN
LN-dc
HN-end
kj-sa
kj-HN
kj-dc"""

lines = """fs-end
he-DX
fs-he
start-DX
pj-DX
end-zg
zg-sl
zg-pj
pj-he
RW-he
fs-DX
pj-RW
zg-RW
start-pj
he-WI
zg-he
pj-fs
start-RW"""

f = open("12.txt")
lines = f.read()
f.close()

lines = lines.split("\n")
lines = [ l.split("-") for l in lines ]
lines

# contruct graph representation
graph = {}
for l in lines:

    if l[0] not in graph: graph[l[0]] = []
    if l[1] not in graph: graph[l[1]] = []

    graph[l[0]].append(l[1])
    graph[l[1]].append(l[0])

graph

# build a dictionary for small and big caves
small = [ node for node in graph if (node not in ["start", "end"]) and (node == node.lower()) ]
small

big = [ node for node in graph if (node not in ["start", "end"]) and (node != node.lower()) ]
big

####################################################################################################
### PART 1

paths = [ ["start"] ]

final = []

# set a really high termination
for path_len in range(1000):

    new_paths = []

    for p in paths:
        for neighbour in graph[p[-1]]:

            new_p = p + [neighbour]
            # print(new_p)

            # remove criterias:

            # putting a start again = error, terminate this path
            if new_p[-1] == "start":
                continue

            # putting an "end" = finalize this path
            if "end" in new_p:
                final.append(new_p)
                continue


            # including a small cave more than once / twice =  error, terminate
            small_error = False

            # part 1 termination
            # small_error = any([1 < new_p.count(char) for char in small])

            # part 2 termination
            visits = [ new_p.count(char) for char in small ]

            # only one 2 present
            if 2 in visits: small_error = (visits.count(2) > 1)
            # exclude 3 visits
            small_error = small_error or (3 in visits)

            if small_error:
                # print("Excluding path:", new_p, "; visits:", visits)
                continue

            new_paths.append(new_p)

    if len(new_paths) == 0:
        print("Cannot generate new path at length", path_len + 1, "\n")
        break
    else:
        # print("length:", path_len + 1, "; paths-in-build:", new_paths, "\n")
        paths = [ new_p.copy() for new_p in new_paths ]

final

ans = len(final)
print("Answer:", ans)

####################################################################################################
### PART 2

# part 2 is the same setup as part 1, only the number of visits to small caves change
# now one small cave can be visited 2 times, and the remaining can be visited at most once
```
